#### Reading the Dataset using pandas read_csv function

```python
import pandas as pd

df= pd.read_csv("tweets_Dataset.csv")

df
```

```python
# Attributes of Dataset before Pre-Processing

df.shape
```

```python
df.size
```

```python
# Get the memory usage of the DataFrame in bytes
memory_usage_bytes = df.memory_usage(deep=True).sum()

# Convert the memory usage to megabytes
memory_usage_MB = memory_usage_bytes / (1024 * 1024)

# Print the memory usage in MB
print(f"Memory usage: {memory_usage_MB:.2f} MB")
```

## Data Pre-Processing

```python
# To Check the null values

# Count the number of NaN values in each row
nan_rows = df.isna().sum(axis=1)
print(nan_rows)

# Count the number of rows with at least one NaN value
num_nan_rows = nan_rows[nan_rows > 0].count()

# Print the result
print(f'The dataset has {num_nan_rows} rows with at least one NaN value.')
```

```python
#Remove the NaN Value

df = df.dropna()

df
```

```python
# To Check the duplicate Rows

duplicate_rows = df.duplicated()

# Print the number of duplicate rows
print(f'The dataset has {duplicate_rows.sum()} duplicate rows.')

# Print the duplicate rows
df[duplicate_rows]
```

```python
# To Remove duplicate rows

df= df.drop_duplicates()

df
```
