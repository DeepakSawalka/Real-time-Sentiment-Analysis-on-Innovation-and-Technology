```python
import pandas as pd
from datetime import datetime, timedelta
from snscrape.modules import twitter

# Define the date range for the tweets
start_date = datetime(2019, 7, 1).date()
end_date = datetime(2019, 8, 30).date()

# Define your query keywords
query_list = ['#ChatGPT', '#BardAI','#Artificial Intelligence','#AI','#BigData','#MachineLearning','#Robotics']

# Define a function to get tweets for a specific date range
def get_tweets_by_date_range(query, start_date, end_date, limit):
    tweets = []
    for tweet in twitter.TwitterSearchScraper(f"{query} since:{start_date} until:{end_date} lang:en").get_items():
        if len(tweets) >= limit:
            break
        tweets.append(tweet)
    return tweets

# Get the tweets and create a dataframe
tweets_list = []
for query in query_list:
    tweets = get_tweets_by_date_range(query, start_date.strftime('%Y-%m-%d'), end_date.strftime('%Y-%m-%d'), 3000)
    for tweet in tweets:
        tweet_dict = {}
        tweet_dict["ID"] = tweet.user.id
        tweet_dict["UserName"] = tweet.user.username
        tweet_dict["Tweets"] = tweet.rawContent
        tweet_dict["Date"] = tweet.date
        tweet_dict["Source"] = tweet.sourceLabel
            
        tweets_list.append(tweet_dict)

df = pd.DataFrame(tweets_list, columns=["ID", "UserName", "Tweets", "Date", "Source"])
df
```

```python
import pandas as pd

# List the names of all 50 CSV files

# Define the file names and the path where the files are located
file_names = ["/content/drive/MyDrive/Tweets_Dataset.csv","/content/drive/MyDrive/dataset/60.csv","/content/drive/MyDrive/dataset/61.csv","/content/drive/MyDrive/dataset/62.csv","/content/drive/MyDrive/dataset/63.csv",
              "/content/drive/MyDrive/dataset/64.csv","/content/drive/MyDrive/dataset/65.csv","/content/drive/MyDrive/dataset/66.csv","/content/drive/MyDrive/dataset/67.csv","/content/drive/MyDrive/dataset/10.csv",
              "/content/drive/MyDrive/dataset/68.csv"]

# Define the chunk size for each file
chunk_size = 1000

# Create an empty DataFrame to store the merged data
merged_file = pd.DataFrame()

# Loop through each CSV file
for file in file_names:
    # Create an empty DataFrame to store the chunk data
    chunk_data = pd.DataFrame()
    
    # Read the CSV file in chunks and append each chunk to the chunk_data DataFrame
    for chunk in pd.read_csv(file, chunksize=chunk_size):
        chunk_data = chunk_data.append(chunk)
    
    # Append the chunk data to the merged_file DataFrame
    merged_file = merged_file.append(chunk_data)



```

```python
# Write the merged data to a new CSV file
merged_file.to_csv("/content/drive/MyDrive/tweets_Dataset.csv", index=False)
```
